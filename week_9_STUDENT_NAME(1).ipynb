Create a bar plot using Seaborn to visualize the count of passengers in each passenger class (Pclass) in the Titanic dataset. Label the x-axis, y-axis, and add a title to the plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('titanic_combined.csv')

# Set the visual style
sns.set(style="whitegrid")

# Create the bar plot
sns.countplot(x='Pclass', data=df)

# Customize labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Class on the Titanic')

# Show the plot
plt.show()
```

Use Plotly to create an interactive scatter plot to visualize the relationship between the age and fare paid by passengers in the Titanic dataset. Include tooltips that display additional information about each data point when hovering over them.

```python
import pandas as pd
import plotly.express as px

# Load your combined dataset
df = pd.read_csv('titanic_combined.csv')

# Drop rows with missing Age or Fare
df_clean = df.dropna(subset=['Age', 'Fare'])

# Create the interactive scatter plot
fig = px.scatter(
    df_clean,
    x='Age',
    y='Fare',
    color='Pclass',  # Optional: color by passenger class
    hover_data=['Name', 'Sex', 'Pclass', 'Survived', 'Source'],  # Tooltip info
    title='Titanic Passengers: Age vs Fare',
    labels={'Age': 'Passenger Age', 'Fare': 'Fare Paid ($)'}
)

# Show the plot
fig.show()
```

Use Seaborn to create a box plot to compare the fare distribution across different passenger classes (Pclass) in the Titanic dataset. Add labels to the x-axis and y-axis, and include a legend to indicate the passenger classes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('titanic_combined.csv')

# Set the visual style
sns.set(style="whitegrid")

# Create the box plot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Pclass', y='Fare', data=df, palette='Set2')

# Customize labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Fare Paid ($)')
plt.title('Fare Distribution by Passenger Class on the Titanic')

# Add legend manually (optional, since Pclass is already on x-axis)

handles, labels = ax.get_legend_handles_labels()
plt.legend(title='Passenger Class', loc='upper right')

# Show the plot
plt.show()
```

Create a line plot using Plotly to display the survival rate of passengers based on their age in the Titanic dataset. Group the data by age range and calculate the survival rate for each group. Add a title and axis labels to the plot

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv('titanic_combined.csv')

# Filter to only rows with known age and survival status
df_filtered = df.dropna(subset=['Age', 'Survived'])

# Convert 'Survived' to numeric (if it's not already)
df_filtered['Survived'] = pd.to_numeric(df_filtered['Survived'], errors='coerce')

# Create age bins (e.g., 09, 1019, ..., 8089)
age_bins = range(0, 90, 10)
df_filtered['AgeGroup'] = pd.cut(df_filtered['Age'], bins=age_bins)

# Calculate survival rate per age group

survival_by_age = df_filtered.groupby('AgeGroup')['Survived'].mean().reset_index()
survival_by_age['AgeGroup'] = survival_by_age['AgeGroup'].astype(str)  # Convert for Plotly

# Create the line plot
fig = px.line(
    survival_by_age,
    x='AgeGroup',
    y='Survived',
    markers=True,
    title='Survival Rate by Age Group on the Titanic',
    labels={'AgeGroup': 'Age Range', 'Survived': 'Survival Rate'}
)

# Show the plot
fig.show()
```

Use Matplotlib to create a stacked bar plot to visualize the survival count of passengers in different passenger classes (Pclass) and embarkation ports (Embarked) in the Titanic dataset. Customize the colors, add a legend, and include a title for the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('titanic_combined.csv')

# Drop rows with missing values in 'Embarked' or 'Survived'
df_clean = df.dropna(subset=['Embarked', 'Survived'])

# Convert 'Survived' to integer if needed
df_clean['Survived'] = df_clean['Survived'].astype(int)

# Group by Pclass and Embarked, then count survivors and non-survivors
grouped = df_clean.groupby(['Pclass', 'Embarked', 'Survived']).size().unstack(fill_value=0)

# Reformat index for plotting
grouped.index = [f'Class {pclass} - Port {port}' for pclass, port in grouped.index]

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
grouped.plot(kind='bar', stacked=True, ax=ax, color=['salmon', 'mediumseagreen'])

# Customize
plt.title('Survival Count by Passenger Class and Embarkation Port')
plt.xlabel('Passenger Class and Embarkation Port')
plt.ylabel('Number of Passengers')
plt.legend(['Did Not Survive', 'Survived'], title='Survival Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Create a heat map using Seaborn to visualize the correlation matrix of numerical variables in the Titanic dataset, such as age, fare, and number of siblings/spouses aboard (SibSp). Customize the color palette and add annotations to the cells to display the correlation values.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('titanic_combined.csv')

# Select numerical columns of interest
numeric_cols = ['Age', 'Fare', 'SibSp']

# Drop rows with missing values in selected columns
df_clean = df[numeric_cols].dropna()

# Compute the correlation matrix
corr_matrix = df_clean.corr()

# Set the visual style
sns.set(style='white')

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,               # Show correlation values
    fmt=".2f",                # Format to 2 decimal places
    cmap='coolwarm',          # Color palette
    vmin=-1, vmax=1,          # Set range for correlation
    linewidths=0.5,           # Add lines between cells
    square=True               # Make cells square
)

# Customize plot
plt.title('Correlation Matrix of Titanic Numerical Features')
plt.tight_layout()
plt.show()
```
