```python
import nltk
from nltk.corpus import gutenberg
from nltk import word_tokenize
from collections import Counter
import math

# Download required resources
nltk.download('gutenberg')
nltk.download('punkt')

# Load and slice texts to avoid memory issues
text1 = gutenberg.raw('melville-moby_dick.txt')[:5000]
text2 = gutenberg.raw('shakespeare-hamlet.txt')[:5000]

# Tokenize and lowercase
tokens1 = word_tokenize(text1.lower())
tokens2 = word_tokenize(text2.lower())

# Create frequency vectors
vec1 = Counter(tokens1)
vec2 = Counter(tokens2)

# Cosine similarity function
def cosine_similarity(c1, c2):
    intersection = set(c1) & set(c2)
    numerator = sum(c1[x] * c2[x] for x in intersection)
    sum1 = sum(v**2 for v in c1.values())
    sum2 = sum(v**2 for v in c2.values())
    denominator = math.sqrt(sum1) * math.sqrt(sum2)
    return numerator / denominator if denominator else 0.0

# Compute and print similarity
cos_sim = cosine_similarity(vec1, vec2)
print("Cosine Similarity between Moby Dick and Hamlet:", cos_sim)
```

```python
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

# Filter tokens
filtered1 = [w.lower() for w in tokens1 if w.isalpha() and w.lower() not in stopwords.words('english')]
filtered2 = [w.lower() for w in tokens2 if w.isalpha() and w.lower() not in stopwords.words('english')]

# Rebuild frequency vectors
vec1 = Counter(filtered1)
vec2 = Counter(filtered2)

# Recompute cosine similarity
cos_sim = cosine_similarity(vec1, vec2)
print("Filtered Cosine Similarity:", cos_sim)
```

```python
import nltk
from nltk.metrics import edit_distance

s1 = "kitten"
s2 = "sitting"

distance = edit_distance(s1, s2)
print("Levenshtein Distance:", distance)

max_len = max(len(s1), len(s2))
similarity = 1 - (distance / max_len)
print("Levenshtein Similarity:", similarity)
```
