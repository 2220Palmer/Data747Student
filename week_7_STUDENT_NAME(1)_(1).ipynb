Visualize the distribution of the target variable using a bar chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Create a bar plot of the 'survived' column
sns.countplot(data=df, x='survived')

# Add labels and title
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Survival on the Titanic')

# Show the plot
plt.show()
```

```python
# Plot histogram of age
sns.histplot(data=df, x='age', bins=30, kde=True, color='skyblue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Age on the Titanic')

# Show the plot
plt.show()
```

Create a scatter plot to visualize the relationship between age and fare

```python
# Create scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='survived', palette='Set1')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare (Colored by Survival)')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Show the plot
plt.show()
```

Identify and visualize any outliers in the dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Use a Box plot for 'age'
sns.boxplot(x=df['age'], color='lightblue')
plt.title('Box Plot of Age')
plt.show()

# Use a Box plot for 'fare'
sns.boxplot(x=df['fare'], color='lightgreen')
plt.title('Box Plot of Fare')
plt.show()
```
