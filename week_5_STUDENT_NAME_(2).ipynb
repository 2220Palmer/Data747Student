1.  Create a line plot of the provided dataset with years on the x-axis and population on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file
df = pd.read_csv("population.csv")

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Population"], marker="o", linestyle="-", color="blue")

# Customize the chart
plt.title("Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.tight_layout()

# Print the chart
plt.show()

```

3. Create a bar chart of the provided dataset with car brands on the x-axis and average price on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("car sales.csv")

plt.figure(figsize=(10, 6))
plt.bar(df["CAR BRAND"], df["AVERAGE PRICE"], color="steelblue")

plt.title("Average Price by Car Brand (2024)")
plt.xlabel("Car Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4. Create a histogram of the provided dataset with salaries on the x-axis and frequency on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("Salary.csv")

plt.figure(figsize=(10, 6))
plt.hist(df["Average Salary"], bins=5, color="skyblue", edgecolor="black")

plt.title("Distribution of Average Salaries (2019–2023)")
plt.xlabel("Average Salary")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()


```

5. Create a pie chart of the provided dataset with the percentage of car sales of each brand.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("car sales.csv")

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    df["% OF SALES"],
    labels=df["CAR BRAND"],
    autopct="%1.1f%%",
    startangle=140,
    colors=["gold", "lightblue", "lightgreen", "salmon", "violet"]
)

plt.title("Market Share by Car Brand (% of Sales in 2024)")
plt.tight_layout()
plt.show()
```

6. Create a stacked bar chart of the provided dataset with the years on the x-axis and the total number of medals won by each country on the y-axis. The stacked bars should show the number of gold, silver, and bronze medals won.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Medals.csv")

df.columns = df.columns.str.strip().str.upper()

# Set index
df.set_index("COUNTRY", inplace=True)

# Plot
df[["GOLD", "SILVER", "BRONZE"]].plot(
    kind="bar",
    stacked=True,
    figsize=(12, 6),
    color=["gold", "silver", "peru"]
)

plt.title("Total Olympic Medals by Country")
plt.xlabel("Country")
plt.ylabel("Number of Medals")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Medal Type")
plt.tight_layout()
plt.show()






```

7. Create a box plot of the provided dataset with car prices grouped by brand.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CAR DATA.csv")

# Plot box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df["PRICE"], vert=False)

plt.title("Box Plot of Car Prices")
plt.xlabel("Price ($)")
plt.tight_layout()
plt.show()
```

8. Create a violin plot of the provided dataset with car prices grouped by brand.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CAR DATA.csv")

# Plot violin plot
plt.figure(figsize=(8, 6))
plt.violinplot(df["PRICE"])

plt.title("Violin Plot of Car Prices")
plt.xlabel("Price ($)")
plt.tight_layout()
plt.show()
```

9. Create a hexbin plot of the provided dataset with car prices on the x-axis and mileage on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CAR DATA.csv")

plt.figure(figsize=(8, 6))
plt.hexbin(df["PRICE"], df["HMPG"], gridsize=10, cmap="Blues", edgecolors="gray")

plt.title("Hexbin Plot: Car Price vs Highway MPG")
plt.xlabel("Price ($)")
plt.ylabel("Highway MPG")
plt.colorbar(label="Count in Bin")
plt.tight_layout()
plt.show()
```

10. Create a 3D scatter plot of the provided dataset with horsepower on the x-axis, weight on the y-axis, and miles per gallon on the z-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv("car sales.csv")

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df["HORSEPOWER"], df["WEIGHT"], df["MPG"], c="teal", s=60)

# Axis labels
ax.set_xlabel("Horsepower")
ax.set_ylabel("Weight (lbs)")
ax.set_zlabel("MPG")

plt.title("3D Scatter Plot: Horsepower vs Weight vs MPG")
plt.tight_layout()
plt.show()

```
